```python
adata_skeletal = sc.read_h5ad('/content/drive/My Drive/Regeneration-order/data/scanpy-objects/WJ.Skeletal.h5ad')
adata_skeletal.X = adata_skeletal.X.astype(np.float64)
print(adata_skeletal)
```

```python
def load_tuple_from_disk(k, address):
    filename =  address + f"tuple_k_{k}.pkl"  # Filename based on k
    with open(filename, 'rb') as file:
        data_tuple = pickle.load(file)
    print(f"Tuple for k={k} loaded from {filename}")
    return data_tuple
```

```python
### address to results of archetypes analyses
address = "/content/drive/My Drive/Regeneration-order/data/tuples-skeletal/"
```

```python
#number of archetypes
for k in range(2,31):
  print(k)
  (similarity_degree_permuted, flattened_perms, X_sorted, model) = load_tuple_from_disk(k,address)
  similarity_df = pd.DataFrame(similarity_degree_permuted)
  similarity_df.index = adata_skeletal.obs[['seurat_clusters']].iloc[flattened_perms].index
  similarity_df.columns = [f'ARCH{i}' for i in range(similarity_df.shape[1])]
  similarity_df.to_csv('/content/drive/My Drive/Regeneration-order/archetypes/data-frames/similarity_df' + "_" + str(k) + '.csv')
```

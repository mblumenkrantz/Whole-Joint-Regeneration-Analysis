```python
import scanpy as sc
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pickle

# Function to save tuple to a file
def save_tuple_to_disk(data_tuple, k, address):
    filename = address + f"tuple_k_{k}.pkl"  # Create a filename based on k
    with open(filename, 'wb') as file:
        pickle.dump(data_tuple, file)
    print(f"Tuple for k={k} saved to {filename}")
```

```python
adata_skeletal = sc.read_h5ad('/content/drive/My Drive/Regeneration-order/data/scanpy-objects/WJ.Skeletal.h5ad')
adata_skeletal.X = adata_skeletal.X.astype(np.float64)
print(adata_skeletal)
```

```python
X = adata_skeletal.obsm['X_pca']
X = pd.DataFrame(X, index=adata_skeletal.obs.index)
X.columns = [f'PC{i+1}' for i in range(X.shape[1])]
X.head()
```

```python
from archetypes import AA
from archetypes.datasets import sort_by_archetype_similarity, permute_dataset
```

```python
def run(df, k=3):

  method_kwargs = {
      "max_iter_optimizer": 10,
  }

  model = AA(n_archetypes=k, method="pgd", method_kwargs=method_kwargs, random_state=3)
  model.fit(df)

  X_sorted, info = sort_by_archetype_similarity(df, [model.similarity_degree_], model.archetypes_)
  similarity_degree_permuted, _ = permute_dataset(model.similarity_degree_, info["perms"])
  flattened_perms = np.concatenate(info["perms"])

  return (similarity_degree_permuted, flattened_perms, X_sorted, model)
```

```python
import os
address = "/content/drive/My Drive/Regeneration-order/data/tuples-skeletal/"
if not os.path.exists(address):
  os.makedirs(address)
```

```python
for k in range(1, 31):
 print(k)
 (similarity_degree_permuted, flattened_perms, X_sorted, model) = run(X, k)
 save_tuple_to_disk((similarity_degree_permuted, flattened_perms, X_sorted, model), k, address)
```
